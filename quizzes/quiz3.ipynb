```python
def is_palindrome(s):
    s = s.lower()
    return s == s[::-1]

def remove_duplicates(lst):
    return list(dict.fromkeys(lst))

from functools import reduce
def factorial(n):
    return reduce(lambda x, y: x * y, range(1, n+1))

def capitalize_first_letter(s):
    return ' '.join(word.capitalize() for word in s.split())

def find_mode(lst):
    return max(set(lst), key=lst.count)

def filter_even(lst):
    return list(filter(lambda x: x % 2 == 0, lst))

def reverse_string(s):
    return s[::-1]

def zip_lists(keys, values):
    return dict(zip(keys, values))

def count_occurrences(lst):
    return {x: lst.count(x) for x in set(lst)}

def strip_whitespace(s):
    return s.strip()

def find_median(lst):
    sorted_lst = sorted(lst)
    n = len(lst)
    if n % 2 == 0:
        return (sorted_lst[n//2 - 1] + sorted_lst[n//2]) / 2
    else:
        return sorted_lst[n//2]

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def square_root(n):
    return n ** 0.5

def sort_descending(lst):
    return sorted(lst, reverse=True)

def list_to_string(lst):
    return ''.join(lst)

def to_lowercase(s):
    return s.lower()

def are_all_unique(lst):
    return len(lst) == len(set(lst))

def sum_list(lst):
    return sum(lst)

def absolute_value(n):
    return abs(n)

if __name__ == "__main__":
    test_string = "A man, a plan, a canal, Panama"
    print("Is palindrome:", is_palindrome(test_string))
    
    test_list = [1, 2, 3, 2, 4, 5, 3, 6, 7]
    print("Unique elements in list:", remove_duplicates(test_list))
    
    print("Factorial of 5:", factorial(5))
    
    test_sentence = "hello world"
    print("Capitalized:", capitalize_first_letter(test_sentence))
    
    print("Mode of list:", find_mode(test_list))
    
    print("Even numbers:", filter_even(test_list))
    
    print("Reversed string:", reverse_string(test_string))
    
    keys = ['a', 'b', 'c']
    values = [1, 2, 3]
    print("Zipped dictionary:", zip_lists(keys, values))
    
    print("Occurrences:", count_occurrences(test_list))
    
    test_string_with_whitespace = "  hello world  "
    print("Stripped:", strip_whitespace(test_string_with_whitespace))
    
    print("Median of list:", find_median(test_list))
    
    print("Is 7 prime?", is_prime(7))
    
    print("Square root of 16:", square_root(16))
    
    print("Sorted descending:", sort_descending(test_list))
    
    test_list_of_strings = ['hello', 'world']
    print("List to string:", list_to_string(test_list_of_strings))
    
    print("Lowercase:", to_lowercase(test_string))
    
    print("All unique?", are_all_unique(test_list))
    
    print("Sum of list:", sum_list(test_list))
    
    print("Absolute value of -5:", absolute_value(-5))
```
