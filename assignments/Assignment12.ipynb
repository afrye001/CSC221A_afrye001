```python
class Money: 
    def __init__(self, amount): 

        self.amount = amount

class Bills(Money): 
    def __init__(self):
        super().__init__(self, amount):

        self.bills = bills

class One(Bills): 
    def __init__(self):
        super().__init__(self, amount):

        self.bill_one = 1.00

class Five(Bills): 
    def __init__(self):
        super().__init__(self, amount):

        self.bill_five = 5.00

class Ten(Bills): 
    def __init__(self):
        super().__init__(self, amount):
 
        self.bill_ten = 10.00

class Twenty(Bills): 
    def __init__(self): 
        super().__init__(self, amount):

        self.bill_twenty = 20.00

class Coins(Money): 
    def __init__(self):
        super().__init__(self, amount):

        self.coins = coins

class Penny(Coins): 
    def __init__(self):
        super().__init__(self, amount):
 
        self.coin_penny = 0.01

class Nickel(Coins): 
    def __init__(self):
        super().__init__(self, amount):
 
        self.coin_nickel = 0.05

class Dime(Coins): 
    def __init__(self): 
        super().__init__(self, amount):

        self.coin_dime = 0.10

class Quarter(Coins): 
    def __init__(self):
        super().__init__(self, amount):

        self.quarter = 0.25 


class Wallet(Money): 
    def __init__(self, money_list):
        super().__init__()(sum(money.amount for money in money_list))
        self.money_list = money_list

        def wallet_info(self): 
            print("Wallet: ") 
            for money in self.money_list: 
                print(" - Type: ", type(money).__name__, "Amount: ", money.amount)
            print("Total amount:", self.amount) 


wallet1 = Wallet([Twenty(1) for _ in range(10)] + [Five(1) for _ in range(2)] + [Quarter(1) for _ in range(3)])
wallet2 = Wallet([Ten(1) for _ in range(5)] + [Dime(1) for _ in range(4)] + [Penny(1) for _ in range(10)])

total_wallet1 = wallet1.amount
total_wallet2 = wallet2.amount 

wallet1.wallet_info()
wallet2.wallet_info()





```
